# Exercice 1 : CartPole

CartPole c'est un jeu où un chariot doit équilibrer un poteau en se déplaçant gauche ou droite.

À chaque pas, l'environnement renvoie 4 valeurs (position du chariot, vitesse, angle du poteau, vitesse angulaire), on choisit une action (0=gauche, 1=droite) et on reçoit +1 de récompense tant que le poteau reste debout.

## Création de l'environnement

```python
import gymnasium as gym

env = gym.make('CartPole-v1')

print(f"Action space: {env.action_space}")
print(f"Sample action: {env.action_space.sample()}")
print(f"Observation space: {env.observation_space}")
print(f"Sample observation: {env.observation_space.sample()}")
```

**Observations :**

- L'espace d'actions est `Discrete(2)` : seulement 2 actions possibles (0=gauche, 1=droite).
- L'espace d'observations est une `Box` avec 4 valeurs continues. Les bornes montrent que la position du chariot est limitée (±4.8) mais que la vitesse et la vitesse angulaire sont théoriquement infinies.

## Politique aléatoire

La politique la plus simple : choisir une action au hasard à chaque pas, sans regarder l'observation.

Elle n'apprend rien. C'est juste un point de départ pour avoir une baseline et voir ensuite à quel point un agent entraîné fait mieux.

```python
for i in range(10):
    obs, info = env.reset()
    episode_over = False
    total_reward = 0

    while not episode_over:
        action = env.action_space.sample()

        observation, reward, terminated, truncated, info = env.step(action)

        total_reward += reward
        episode_over = terminated or truncated

    print(f"Episode {i+1} finished! Total reward: {total_reward}")
env.close()
```

## Analyse des résultats

Les scores varient beaucoup d'un épisode à l'autre, ce qui est normal puisque tout dépend du hasard. On tourne autour de 20-25 pas en moyenne alors que le max possible est 500.

C'est clairement une très mauvaise baseline. Un agent entraîné (Q-Learning, DQN) devra largement dépasser ça pour qu'on puisse dire qu'il a vraiment appris quelque chose.
