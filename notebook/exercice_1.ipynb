# Exercice 1 : CartPole


CartPole est un jeu où un chariot se déplace sur un rail et doit équilibrer un poteau en se déplaçant gauche ou droite. 

À chaque instant, l'environnement te donne une observation (position du chariot, vitesse, angle du poteau, vitesse angulaire), tu choisis une action (gauche=0 ou droite=1), et tu reçois une récompense (+1 pour chaque pas où le poteau reste debout).

## Création de l'environnement

```python
import gymnasium as gym

env = gym.make('CartPole-v1')

print(f"Action space: {env.action_space}")  # Discrete(2) - left or right
print(f"Sample action: {env.action_space.sample()}")  # 0 or 1

# Box observation space (continuous values)
print(f"Observation space: {env.observation_space}")  # Box with 4 values
# Box([-4.8, -inf, -0.418, -inf], [4.8, inf, 0.418, inf])
print(f"Sample observation: {env.observation_space.sample()}")  # Random valid observation
```

## Politique aléatoire

Une politique définit comment un agent choisit ses actions. La politique la plus simple possible est la politique aléatoire : à chaque pas, l'agent choisit une action au hasard, sans tenir compte de l'observation reçue.

C'est notre point de départ. Elle ne "apprend" rien — elle sert de baseline pour mesurer à quel point un agent intelligent fera mieux.

```python


for i in range(10):

    obs, info = env.reset()  # Start properly
    episode_over = False
    total_reward = 0

    while not episode_over:
        # Choose an action: 0 = push cart left, 1 = push cart right
        action = env.action_space.sample()  # Random action for now - real agents will be smarter!

        observation, reward, terminated, truncated, info = env.step(action)

        total_reward += reward
        episode_over = terminated or truncated

    print(f"Episode {i+1} finished! Total reward: {total_reward}")
env.close()
```

## Analyse des résultats

Les scores varient fortement d'un épisode à l'autre (entre ~8 et ~60 pas), ce qui est attendu : sans apprentissage, chaque épisode dépend uniquement de la chance. La moyenne tourne autour de 20-25 pas sur CartPole-v1, alors que le score maximum possible est 500.

Ces résultats confirment l'intuition : une politique aléatoire est une baseline très faible. Un agent entraîné (par Q-Learning ou DQN) devra dépasser largement ces scores pour prouver qu'il a appris quelque chose.
